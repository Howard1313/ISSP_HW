```python
# Reference:https://towardsdatascience.com/quantum-physics-visualization-with-python-35df8b365ff
import matplotlib.pyplot as plt
import numpy as np
```

```python
def prob_2s(x,y,z):
    r=np.sqrt(np.square(x)+np.square(y)+np.square(z))
    return np.square((2-r)*np.exp(-r/2))

def prob_2pz(x,y,z):
    r=np.sqrt(np.square(x)+np.square(y)+np.square(z))
    theta=np.arccos(z/r)
    phi=np.arctan(y/x)
    return np.square(r*np.exp(-r/2)*np.cos(theta))
def prob_2px(x,y,z):
    r=np.sqrt(np.square(x)+np.square(y)+np.square(z))
    theta=np.arccos(z/r)
    phi=np.arctan(y/x)
    return np.square(r*np.exp(-r/2)*np.sin(theta)*np.cos(phi))
def prob_2py(x,y,z):
    r=np.sqrt(np.square(x)+np.square(y)+np.square(z))
    theta=np.arccos(z/r)
    phi=np.arctan(y/x)
    return np.square(r*np.exp(-r/2)*np.sin(theta)*np.sin(phi))
def prob_2py(x,y,z):
    r=np.sqrt(np.square(x)+np.square(y)+np.square(z))
    theta=np.arccos(z/r)
    phi=np.arctan(y/x)
    return np.square(r*np.exp(-r/2)*np.sin(theta)*np.sin(phi))

def prob_3d1(x,y,z):
    r=np.sqrt(np.square(x)+np.square(y)+np.square(z))
    theta=np.arccos(z/r)
    phi=np.arctan(y/x)
    return np.square(r*r*np.exp(-r/2)*(3*np.cos(theta)*np.cos(theta)-1))
def prob_3d2(x,y,z):
    r=np.sqrt(np.square(x)+np.square(y)+np.square(z))
    theta=np.arccos(z/r)
    phi=np.arctan(y/x)
    return np.square(r*r*np.exp(-r/2)*(np.sin(theta)*np.cos(theta)*np.cos(phi)))
def prob_3d3(x,y,z):
    r=np.sqrt(np.square(x)+np.square(y)+np.square(z))
    theta=np.arccos(z/r)
    phi=np.arctan(y/x)
    return np.square(r*r*np.exp(-r/2)*(np.sin(theta)*np.cos(theta)*np.sin(phi)))
def prob_3d4(x,y,z):
    r=np.sqrt(np.square(x)+np.square(y)+np.square(z))
    theta=np.arccos(z/r)
    phi=np.arctan(y/x)
    return np.square(r*r*np.exp(-r/2)*(np.sin(theta)*np.sin(theta)*np.cos(2*phi)))
def prob_3d5(x,y,z):
    r=np.sqrt(np.square(x)+np.square(y)+np.square(z))
    theta=np.arccos(z/r)
    phi=np.arctan(y/x)
    return np.square(r*r*np.exp(-r/2)*(np.sin(theta)*np.sin(theta)*np.sin(2*phi)))
```

```python
#Random coordinates
x=np.linspace(-6,6,30)
y=np.linspace(-6,6,30)
z=np.linspace(-6,6,30)
elements = []
probability = []
for ix in x:
    for iy in y:
        for iz in z:
            #Serialize into 1D object
            elements.append(str((ix,iy,iz)))
            probability.append(prob_3d1(ix,iy,iz))
            
#Ensure sum of probability is 1
probability = probability/sum(probability)
#Getting electron coordinates based on probabiliy
coord = np.random.choice(elements, size=100000, replace=True, p=probability)
# print(coord[0:10])
elem_mat = [i.split(',') for i in coord]
# print(elem_mat[0:10])
# print(len(elem_mat))
elem_mat = np.array(elem_mat)
# print(elem_mat[0:10])
# print(elem_mat.shape)
x_coords = [float(i.item()[1:]) for i in elem_mat[:,0]] 
y_coords = [float(i.item()) for i in elem_mat[:,1]] 
z_coords = [float(i.item()[0:-1]) for i in elem_mat[:,2]]
#Plotting
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_coords, y_coords, z_coords, alpha=0.05, s=2)
ax.set_title("Hydrogen density")
plt.show()
```
